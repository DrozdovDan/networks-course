```python
import matplotlib.pyplot as plt
import numpy as np

N_values = np.array([10, 100, 1000])
d = 2000
u_values = np.array([300, 700, 2000])
u_s = 30000
F = 15 * 10**6

def client_server(F, d, u_s, N):
    return np.maximum(N * F/ u_s, F / d)

def p_2_p(F, d, u_s, N, u):
    return np.maximum(max(F / u_s, F / d), N * F / (u_s + N * u))

plt.figure(figsize=(8,6))

plt.plot(N_values, client_server(F, d, u_s, N_values), marker='o', label='Client-Server')
plt.plot(N_values, p_2_p(F, d, u_s, N_values, u_values[0]), marker='s', label='P2P, u = 300 Kbit/s')
plt.plot(N_values, p_2_p(F, d, u_s, N_values, u_values[1]), marker='^', label='P2P, u = 700 Kbit/s')
plt.plot(N_values, p_2_p(F, d, u_s, N_values, u_values[2]), marker='d', label='P2P, u = 2 Mbit/s')

plt.xscale('log')
plt.yscale('log')

plt.xlabel('Количество пирам (N)')
plt.ylabel('Минимальное время раздачи (с)')
plt.title('Сравнение минимального времени раздачи\nдля клиент-серверной и P2P раздачи')
plt.legend()
plt.grid(True, which="both", ls="--")
plt.show()
```
